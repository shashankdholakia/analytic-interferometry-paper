```python
import paths
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv(paths.data / 'asu.tsv', sep='\t', index_col=0, on_bad_lines='warn', comment='#', skiprows=[0,1])
```

```python
df = df[2:]
```

```python
vmag = df['Vmag'].astype(float)
ldd = df['LDD'].astype(float)
eldd = df['e_LDD'].astype(float)
plt.errorbar(vmag, ldd, yerr=eldd, fmt='o')
plt.ylim(0,3)
```
